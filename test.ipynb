```python
import pandas as pd

# Load the dataset to analyze its contents
file_path = 'store_data_2023.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Identify string (categorical) features to encode
categorical_features = data.select_dtypes(include=['object']).columns

```

```python
# Drop non-informative string features like 'date' or others irrelevant to turnover
irrelevant_features = ['date', 'postal', 'street']
categorical_features = [col for col in categorical_features if col not in irrelevant_features]
```

```python
# One-hot encode categorical features
data_encoded = pd.get_dummies(data, columns=categorical_features, drop_first=True)
```

```python
# Drop irrelevant features from the encoded dataset
data_encoded.drop(columns=irrelevant_features, inplace=True, errors='ignore')
```

```python
# Calculate the correlation matrix
correlation_matrix = data_encoded.corr()
```

```python
# Extract correlations with the 'turnover' target feature
turnover_correlations = correlation_matrix['turnover'].sort_values(ascending=False)

# Display the correlation matrix for 'turnover'
turnover_correlations.head(10), turnover_correlations.tail(10)
```
